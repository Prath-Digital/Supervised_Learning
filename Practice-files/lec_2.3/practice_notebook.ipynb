```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
house_price_data = {
    'house_sqr_ft': [1200, 1500, 1800, 900, 2000],
    'price': [200000, 250000, 300000, 160000, 330000]
}

df = pd.DataFrame(house_price_data)
df
```

```python
lr_model = LinearRegression()
lr_model.fit(df[['house_sqr_ft']], df['price'])
```

```python
prediction = lr_model.predict([[1000]])
print(prediction)
```

```python
prediction = lr_model.predict(df[['house_sqr_ft']])
print(prediction)
```

```python
slope = lr_model.coef_[0]
print(slope)
```

```python
intercept = lr_model.intercept_
print(intercept)
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(df[['price']], lr_model.predict(df[['house_sqr_ft']]))
print(r2)
```

```python
df = pd.read_csv("placement.csv")
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
lr_model = LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['cgpa'], df['package'], test_size=0.2, random_state=40)
```

```python
X_train = pd.DataFrame(X_train)
X_train
```

```python
lr_model.fit(X_train, y_train)
```

```python
X_test = pd.DataFrame(X_test)
```

```python
y_pred = lr_model.predict(X_test)
y_pred
```

```python
X_test = X_test.values.flatten()
```

```python
X_test.shape
```

```python
y_pred.shape
```

```python
final_df = pd.DataFrame({'cgpa_test': X_test, 'package_pred': y_pred})
final_df
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2 = r2_score(y_test, y_pred)
print(f"R2 Score: {r2}")
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(f"MAE: {mae} LPA")
print(f"MSE: {mse}")
print(f"RMSE: {mse ** 0.5} LPA")
```

```python
n = X_train.shape[0]
p = 1
```

```python
adj_r2 = 1 - (((1 - r2) * (n - 1)) / (n-p-1))

print(f"Adj R2 Score: {adj_r2}")
```
