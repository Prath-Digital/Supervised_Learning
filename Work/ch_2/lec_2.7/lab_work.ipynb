```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

# Q1 Basic Gradient Descent (Single Variable)

```python
theta=0
lr=0.1

print('Iteration | Theta')

for i in range(20):
    gradient=2*(theta-5)
    theta=theta-lr*gradient
    print(i+1,'|',theta)
```

# Q2 Gradient Descent for Linear Regression

```python
X=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])

b0=0
b1=0
lr=0.01
n=len(X)

for i in range(50):
    y_pred=b0+b1*X

    db0=(-2/n)*sum(y-y_pred)
    db1=(-2/n)*sum(X*(y-y_pred))

    b0=b0-lr*db0
    b1=b1-lr*db1

print('Intercept=',b0)
print('Slope=',b1)
```

# Q3 Cost Function Tracking

```python
b0=0
b1=0
costs=[]

for i in range(50):

    y_pred=b0+b1*X

    cost=np.mean((y-y_pred)**2)
    costs.append(cost)

    db0=(-2/n)*sum(y-y_pred)
    db1=(-2/n)*sum(X*(y-y_pred))

    b0=b0-lr*db0
    b1=b1-lr*db1

plt.plot(costs)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost vs Iterations')
plt.show()
```

# Q4 Learning Rate Experiment

```python
rates=[0.001,0.01,0.1]

for lr in rates:

    b0=0
    b1=0
    costs=[]

    for i in range(50):

        y_pred=b0+b1*X
        cost=np.mean((y-y_pred)**2)
        costs.append(cost)

        db0=(-2/n)*sum(y-y_pred)
        db1=(-2/n)*sum(X*(y-y_pred))

        b0=b0-lr*db0
        b1=b1-lr*db1

    plt.plot(costs,label='LR='+str(lr))

plt.legend()
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()
```

# Q5 Batch Gradient Descent

```python
X=np.array([1,2,3,4,5])
y=np.array([3,5,7,9,11])

b0=0
b1=0
lr=0.01

for i in range(50):

 y_pred=b0+b1*X

 db0=(-2/n)*sum(y-y_pred)
 db1=(-2/n)*sum(X*(y-y_pred))

 b0=b0-lr*db0
 b1=b1-lr*db1

print('Intercept=',b0)
print('Slope=',b1)
```

# Q6 Compare Predictions

```python
pred=b0+b1*X

print('Actual:',y)
print('Predicted:',pred)
```
