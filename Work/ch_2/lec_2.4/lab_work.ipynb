# Q1 Understanding the MLR Equation

```python
# Mathematical Equation:
# y = β0 + β1x1 + β2x2

print("y = β0 + β1x1 + β2x2")

print("\nVariable Mapping:")
print("x1 = Experience")
print("x2 = Education Level")
print("y = Salary")

experience = 5
education = 3

b0 = 5000
b1 = 2000
b2 = 1500

salary = b0 + b1 * experience + b2 * education

print("\nSample Prediction=", salary)
```

# Q3 Manual Prediction Using Coefficients

```python
b0 = 5000
b1 = 2000
b2 = 1500

samples = np.array([[2, 1], [5, 2], [7, 4]])

print("Manual Predictions:")

for x1, x2 in samples:
    pred = b0 + b1 * x1 + b2 * x2
    print((x1, x2), "=", pred)

X = np.column_stack((np.ones(3), samples))

beta = np.array([b0, b1, b2])

prediction = X @ beta

print("\nNumPy Verification:")
print(prediction)
```
