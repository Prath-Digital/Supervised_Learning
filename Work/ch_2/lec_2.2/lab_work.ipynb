# Q.1 — Compute SSE Manually

Given:

- $x = [1, 2, 3]$
- $y = [2, 4, 5]$
- Regression Line:

$$
\hat{y} = 1 + 1.2x
$$

Steps:
1. Compute predicted values
2. Compute residuals
3. Compute SSE

```python
import numpy as np

# Given data
x = np.array([1, 2, 3])
y = np.array([2, 4, 5])

# Regression equation
y_hat = 1 + 1.2 * x

# Residuals
errors = y - y_hat

# SSE
SSE = np.sum(errors ** 2)

print('Predicted Values:', y_hat)
print('Residuals:', errors)
print('SSE:', SSE)
```

# Q.2 — Derivative Calculation

Using SSE formula:

$$
SSE = \sum (y_i - \beta_0 - \beta_1 x_i)^2
$$

Find:

$$
\frac{\partial SSE}{\partial \beta_0}
$$

$$
\frac{\partial SSE}{\partial \beta_1}
$$


## Partial Derivative with respect to β₀

$$
SSE = \sum (y_i - \beta_0 - \beta_1 x_i)^2
$$

Differentiate:

$$
\frac{\partial SSE}{\partial \beta_0}
$$

$$
= \sum 2(y_i - \beta_0 - \beta_1x_i)(-1)
$$

$$
= -2\sum (y_i - \beta_0 - \beta_1x_i)
$$

---

## Partial Derivative with respect to β₁

$$
\frac{\partial SSE}{\partial \beta_1}
$$

$$
= \sum 2(y_i - \beta_0 - \beta_1x_i)(-x_i)
$$

$$
= -2\sum x_i(y_i - \beta_0 - \beta_1x_i)
$$


# Q.3 — Matrix Form Exercise

Given:

$$
X =
\begin{bmatrix}
1 & 2 \\
1 & 3 \\
1 & 4
\end{bmatrix}
$$

$$
y =
\begin{bmatrix}
4 \\
6 \\
8
\end{bmatrix}
$$

Compute:
1. $X^TX$
2. $X^Ty$
3. $(X^TX)^{-1}$
4. $\hat{\beta} = (X^TX)^{-1}X^Ty$

```python
import numpy as np

# Matrix X
X = np.array([
    [1, 2],
    [1, 3],
    [1, 4]
])

# Vector y
y = np.array([
    [4],
    [6],
    [8]
])

# Calculations
XT_X = X.T @ X
XT_y = X.T @ y
inverse = np.linalg.inv(XT_X)
beta_hat = inverse @ XT_y

print('X^T X =')
print(XT_X)

print('\nX^T y =')
print(XT_y)

print('\n(X^T X)^-1 =')
print(inverse)

print('\nBeta Hat =')
print(beta_hat)
```

# Q.4 — Plot Cost Function

Tasks:
1. Vary β₁ from -5 to 5
2. Compute SSE for each value
3. Plot SSE vs β₁
4. Observe minimum error point

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
x = np.array([1, 2, 3])
y = np.array([2, 4, 5])

# Fixed intercept
beta_0 = 1

# Range of beta_1 values
beta_1_values = np.linspace(-5, 5, 100)

sse_values = []

# Compute SSE for each beta_1
for beta_1 in beta_1_values:
    y_hat = beta_0 + beta_1 * x
    sse = np.sum((y - y_hat) ** 2)
    sse_values.append(sse)

# Plot
plt.figure(figsize=(10,6))
plt.plot(beta_1_values, sse_values)
plt.xlabel('Beta_1')
plt.ylabel('SSE')
plt.title('SSE vs Beta_1')
plt.grid(True)
plt.show()
```
