## Given Dataset

- x = [1, 2, 3]
- y = [2, 4, 5]


```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

# Q1 - Manual Computation of $\beta_0$ and $\beta_1$

```python
# Given data
x = np.array([1, 2, 3])
y = np.array([2, 4, 5])

# Step 1: Compute means
x_mean = np.mean(x)
y_mean = np.mean(y)

print('Mean of x =', x_mean)
print('Mean of y =', y_mean)
```

## Formula for β₁

$$
\beta_1 = \frac{\sum{(x - \bar{x})(y - \bar{y})}}{\sum{(x - \bar{x})^2}}
$$

```python
# Step 2: Compute numerator
numerator = np.sum((x - x_mean) * (y - y_mean))

# Step 3: Compute denominator
denominator = np.sum((x - x_mean) ** 2)

print('Numerator =', numerator)
print('Denominator =', denominator)
```

## Compute $\beta_1$ and $\beta_0$

```python
# Step 4: Compute beta values
beta1 = numerator / denominator
beta0 = y_mean - beta1 * x_mean

print('β₁ (Slope) =', beta1)
print('β₀ (Intercept) =', beta0)
```

## Final Regression Equation

$$
\hat{y} = \beta_0 + \beta_1x
$$

```python
print(f'Regression Equation: ŷ = {beta0:.2f} + {beta1:.2f}x')
```

# Q2 - Compute Predictions

```python
# Predicted values
y_hat = beta0 + beta1 * x

print('Predicted values:')
print(y_hat)
```

## Compute Residuals

```python
# Residuals
e = y - y_hat

print('Residuals:')
print(e)
```

## Compute SSE (Sum of Squared Errors)

$$
SSE = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

```python
# Compute SSE
SSE = np.sum(np.square(e))

print('SSE =', SSE)
```

# Q3 - Plot the Regression Line

```python
plt.figure(figsize=(8,5))

# Scatter plot
plt.scatter(x, y, label='Actual Data Points', s=100)

# Regression line
plt.plot(x, y_hat, label='Regression Line')

# Labels and title
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Linear Regression Line')

plt.legend()
plt.grid(True)
plt.show()
```

# Q4 - Explain in Your Own Words

## What $\beta_0$ Represents

$\beta_0$ is called the intercept. It represents the predicted value of y when x = 0.

---

## What $\beta_1$ Represents

$\beta_1$ is called the slope. It represents how much y changes when x increases by 1 unit.

---

## What SSE Represents

SSE (Sum of Squared Errors) measures the total error between actual values and predicted values.

Smaller SSE means the regression line fits the data better.

---

## Why Linear Regression Minimizes SSE

Linear Regression chooses the best regression line by minimizing the SSE value.

This helps the model make predictions as close as possible to the actual data points.
