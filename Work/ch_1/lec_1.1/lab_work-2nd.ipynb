## Q.3. Choose any one problem (house price prediction, student marks prediction, fraud detection).
Write down:
- Problem Statement
- Data sources
- Preprocessing steps
- Model type
- How to evaluate the model

### House Price Prediction 🏠
#### Problem Statement
Predict the price of a house using features like location, size, and number of rooms.

#### Data Sources
- Real estate websites
- Housing datasets
- Government property records

#### Preprocessing Steps
- Remove missing data
- Convert text data into numbers
- Scale numerical values
- Split data into training and testing sets

#### Model Type
**Linear Regression** is commonly used for predicting house prices.

#### Model Evaluation
- **MAE** – average error
- **MSE** – measures prediction error
- **R² Score** – checks model accuracy
