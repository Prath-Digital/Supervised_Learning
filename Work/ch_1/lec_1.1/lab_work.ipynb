## Q.6. Prepare a short explanation of:
"How is Machine Learning transforming the world?"
Share examples and discuss impact.

Machine Learning (ML) is transforming the world by enabling computers to learn from data and make decisions without being explicitly programmed. This has led to significant advancements across various industries, improving efficiency, accuracy, and personalization.

In healthcare, ML is being used for early disease detection, personalized treatment plans, and drug discovery. For instance, ML algorithms can analyze medical images to identify tumors or predict patient outcomes, leading to better diagnosis and treatment.

In finance, ML is revolutionizing fraud detection, risk assessment, and algorithmic trading. Banks and financial institutions use ML models to analyze transaction patterns and identify fraudulent activities in real-time, enhancing security and trust.

detection).

- Problem Statement
- Data sources
- Preprocessing steps
- Model type
- How to evaluate the model

### House Price Prediction 🏠
#### Problem Statement
Predict the price of a house using features like location, size, and number of rooms.

#### Data Sources
- Real estate websites
- Housing datasets
- Government property records

#### Preprocessing Steps
- Remove missing data
- Convert text data into numbers
- Scale numerical values
- Split data into training and testing sets

#### Model Type
**Linear Regression** is commonly used for predicting house prices.

#### Model Evaluation
- **MAE** – average error
- **MSE** – measures prediction error
- **R² Score** – checks model accuracy
