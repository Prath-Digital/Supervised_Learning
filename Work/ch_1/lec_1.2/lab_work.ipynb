## Q.1. Supervised learning solves two types of tasks:
1. Regression - Predict continuous numeric values
2. Classification - Predict discrete class labels

List all Common Supervised Algorithms of the above types.

---

1. Common Supervised Algorithms for Regression:
    - Linear Regression
    - Polynomial Regression
    - Support Vector Regression
    - Random Forest Regression
    - Gradient Boosting Regression (e.g., XGBoost)

2. Common Supervised Algorithms for Classification:
    - Logistic Regression
    - Decision Trees
    - Random Forest
    - Support Vector Machines
    - K-Nearest Neighbors

## Q.2. For each supervised algorithm type (Regression, Classification), list three real-world applications.

---

1. Real-world applications of Regression:
    - Predicting house prices based on features like size, location, and number of bedrooms.
    - Forecasting stock prices based on historical data and market indicators.
    - Estimating the amount of rainfall based on weather patterns and atmospheric conditions.
2. Real-world applications of Classification:
    - Email spam detection, classifying emails as spam or not spam.
    - Medical diagnosis, classifying patients as having a certain disease or not based on symptoms and test results.
    - Animal recognition, classifying images of animals into different species based on visual features.
